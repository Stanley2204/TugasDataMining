```python
# Breast Cancer Dataset Analysis

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Load the dataset
breast_cancer_df = pd.read_csv('https://raw.githubusercontent.com/Stanley2204/TugasDataMining/refs/heads/main/breast-cancer.csv')
breast_cancer_df.head()
```

```python
# Data Preprocessing
breast_cancer_df.drop('id', axis=1, inplace=True)
breast_cancer_df['diagnosis'] = breast_cancer_df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# Splitting the dataset
X_bc = breast_cancer_df.drop("diagnosis", axis=1)
y_bc = breast_cancer_df["diagnosis"]
scaler = StandardScaler()
X_bc_scaled = scaler.fit_transform(X_bc)
X_bc_train, X_bc_test, y_bc_train, y_bc_test = train_test_split(X_bc_scaled, y_bc, test_size=0.2, random_state=42)
```

```python
# Classification Models
dt_bc_clf = DecisionTreeClassifier(random_state=42)
dt_bc_clf.fit(X_bc_train, y_bc_train)
y_bc_pred_dt = dt_bc_clf.predict(X_bc_test)

rf_bc_clf = RandomForestClassifier(random_state=42)
rf_bc_clf.fit(X_bc_train, y_bc_train)
y_bc_pred_rf = rf_bc_clf.predict(X_bc_test)

svm_bc_clf = SVC(random_state=42)
svm_bc_clf.fit(X_bc_train, y_bc_train)
y_bc_pred_svm = svm_bc_clf.predict(X_bc_test)

# Evaluation
print('Decision Tree Accuracy:', accuracy_score(y_bc_test, y_bc_pred_dt))
print('Random Forest Accuracy:', accuracy_score(y_bc_test, y_bc_pred_rf))
print('SVM Accuracy:', accuracy_score(y_bc_test, y_bc_pred_svm))

print('Decision Tree Classification Report:\n', classification_report(y_bc_test, y_bc_pred_dt))
print('Random Forest Classification Report:\n', classification_report(y_bc_test, y_bc_pred_rf))
print('SVM Classification Report:\n', classification_report(y_bc_test, y_bc_pred_svm))
```
